```python
import numpy as np
import samplebase as sb
import path_utils as putils
```

```python
data_dir = putils.data_dir(18)
data_dir
```

timestep `dt` is not set here anymore since trajectory is created in a different place. The same goes for `target_time`.

```python
alphas = np.logspace(-8, -2, num=50) # np.logspace(-9, -3, num=64)
gillespie_realisations = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000]
n_splitses = [10]#, 10, 15]#[4, 7, 15, 20, 30]
n_repetitions = 10
print(alphas)
print(n_splitses)
print(gillespie_realisations)
print(n_repetitions)
print("number of samples = {}".format(n_repetitions * len(alphas) * len(n_splitses) * len(gillespie_realisations)))
```

```python
n_splits = n_splitses[0]
batch_size = 10
for a in [alphas[i:i + batch_size] for i in range(0, len(alphas), batch_size)]:
    for gr in gillespie_realisations:
        for i in range(n_repetitions):
            args = {
                "alpha": a,
                "n_splits": n_splits,
                "gillespie_realisations": gr,
                "iid_id": i,
                "traj_file_path": "./gillespie_two_inits_conced.h5"
            }
            sb.create_sample(data_dir, args=args)
        print("Created jobs for alpha batch {}".format(a))
```
